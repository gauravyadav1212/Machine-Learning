```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
```

```python
df
```

```python
target = df['salary_more_then_100k']
inputs = df.drop(['salary_more_then_100k'], axis = 1)
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
inputs['company'] = le.fit_transform(df['company'])
inputs['job'] = le.fit_transform(df['job'])
inputs['degree'] = le.fit_transform(df['degree'])
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))  # Set the figure size
tree.plot_tree(model, feature_names=['company', 'job', 'degree'], class_names=['0', '1'], filled=True)
plt.show()
```
